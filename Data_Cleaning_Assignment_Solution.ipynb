# Data Cleaning Assignment Solution
Dataset: healthcare_data_cleaning_dataset.csv

```

import pandas as pd
import numpy as np

df = pd.read_csv('healthcare_data_cleaning_dataset.csv')
df.head()
```

## Q1. Missing Data Identification

```

missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum()/len(df))*100

print("Missing Values:")
print(missing_values)

print("\nMissing Percentage:")
print(missing_percentage.round(2))
```

## Q2. Handling Missing Age
Median is used because it is robust to extreme values.

```

age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)
print("Median Age:", age_median)
```

## Q3. Handling Missing Treatment Cost
Median is used because Treatment Cost is skewed.

```

cost_median = df['Treatment_Cost'].median()
df['Treatment_Cost'] = df['Treatment_Cost'].fillna(cost_median)
print("Median Treatment Cost:", cost_median)
```

## Q4. Duplicate Patient Records

```

before_rows = len(df)
duplicates = df.duplicated().sum()

df = df.drop_duplicates()

after_rows = len(df)

print("Duplicate Rows:", duplicates)
print("Before:", before_rows)
print("After:", after_rows)
```

## Q5. Invalid Age Values

```

invalid_age = df[(df['Age'] < 0) | (df['Age'] > 100)]
print("Invalid Age Records:", len(invalid_age))
invalid_age.head()
```

## Q6. Outlier Detection using IQR

```

Q1 = df['Treatment_Cost'].quantile(0.25)
Q3 = df['Treatment_Cost'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df['Treatment_Cost'] < lower) | (df['Treatment_Cost'] > upper)]
print("Number of Outliers:", len(outliers))
```

## Q7. Outlier Treatment (Winsorization)

```

p5 = df['Treatment_Cost'].quantile(0.05)
p95 = df['Treatment_Cost'].quantile(0.95)

df['Treatment_Cost_Capped'] = df['Treatment_Cost'].clip(lower=p5, upper=p95)

print("5th Percentile:", p5)
print("95th Percentile:", p95)
```

## Q8. Log Transformation

```

df['Treatment_Cost_Log'] = np.log1p(df['Treatment_Cost'])

df[['Treatment_Cost','Treatment_Cost_Log']].head()
```

## Q9. Time-Based Missing Handling

```

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

df = df.sort_values('Admission_Date')

df = df.ffill()

print("Dataset sorted by Admission_Date and forward fill applied.")
```
